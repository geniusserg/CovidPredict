### Data preparation 

### Scaling and imputing

Scale values as (x-min(x))/(max(x)-min(x))

Impute NaN values using MICE method with IterativeImputer on Linear regression as estimator 

```python
#Function which prepare data
class DataPreprocessor():
    def __init__(self, scal_model=StandardScaler(), imputer=IterativeImputer(max_iter=50), window=1):
        self.scal_model = scal_model
        self.imputer = imputer
        self.window = window
        
    def prepare_data(self, df_exp,  window=1):
        df_exp = df_exp.loc[(df_exp.groupby("case").size()>1).index]
        df_exp = df_exp.drop(["current_process_duration", "admission_date", "end_epizode", "outcome_tar", "long_observation_tar", "process_stages"], axis=1)
        df_exp.iloc[:, 29:42] = self.scal_model.fit_transform(df_exp.iloc[:, 29:42])
        df_exp.iloc[:, 29:42] = self.imputer.fit_transform(df_exp.iloc[:, 29:42])
        return df_exp
    
    def reverse_data(self, x):
        return self.scal_model.inverse_transform(x)

df_exp = DataPreprocessor().prepare_data(df)
df_exp
```

### Function which shows results of model

```python
# Function which shows results of model and saves
def show_results(history, table):
    plt.plot(history.history["loss"])
    plt.plot(history.history["val_loss"])
    plt.legend(["train loss", "validation loss"])
    window = history.history["window"]
    model = history.history["model"]
    plt.title(f"Model: {model.name}, window: {window}")
    df_score = pd.DataFrame(table, index=[i[:-11] for i in df.columns[29:42]])
    df_score["MAPE"] = df_score["MAPE"].map(lambda x: f"{(x*100):.2f}"+"%")
    df_score.style.set_caption(f"Validation metrics. Model: {model.name}, window: {window}")
    model_report_dir = f"{report_path}/report_w{window}"
    if not os.path.exists(model_report_dir):
        os.mkdir(model_report_dir)
        
    df_score.to_csv(model_report_dir+"/"+model.name+".csv")
    display(df_score)
    plt.show()
```

### Function what trains model

```python
GLOB_RR_TABLE = []
class ModelTrainer():
    def __init__(self, preprocessor, df_exp):
        self.preprocessor = preprocessor
        self.df_exp = df_exp

    def __prepare_data(self):
        X, y = [], []
        window = self.preprocessor.window
        for i in self.df_exp.index.levels[0]:
            for j in range(len(self.df_exp.loc[i])-window):
                X.append(self.df_exp.loc[(i, j): (i, j+window-1)].to_numpy())
                y.append(self.df_exp.loc[(i, j+window)].iloc[29:42].to_numpy())
        return np.array(X), np.array(y)
        
    #Function which train model
    def train_model(self, model):
        X, y = self.__prepare_data()
        kfold = KFold(n_splits=5)
        table = {"MSE":[], "MAPE":[], "R2":[]}
        table_avg = {"MSE":[], "MAPE":[], "R2":[]}
        history = None
        for i, (train_index, test_index) in enumerate(kfold.split(X)):
            history = model.fit(X[train_index], y[train_index], validation_split=0.1, epochs = 30, verbose=0)

            y_pred = model.predict(X[test_index], verbose=0)
            y_pred_inv = self.preprocessor.scal_model.inverse_transform(y_pred)
            y_real_inv = self.preprocessor.scal_model.inverse_transform(y[test_index])

            rsq_score = r2_score(y[test_index], y_pred, multioutput="raw_values")
            mse_score = mean_squared_error(y[test_index], y_pred, multioutput="raw_values")
            mape_score = mean_absolute_percentage_error(y_real_inv, y_pred_inv, multioutput="raw_values")
            
            table["MSE"].append(mse_score)
            table["MAPE"].append(mape_score)
            table["R2"].append(rsq_score)
        
        for metric in table:
            cv_num = len(table[metric])
            param_num = len(table[metric][0])
            for l in range(param_num):
                table_avg[metric].append(np.mean([table[metric][k][l] for k in range(cv_num)]))
    
        
        history.history["window"] = self.preprocessor.window
        history.history["model"] = model
        
        return history, table_avg

```
